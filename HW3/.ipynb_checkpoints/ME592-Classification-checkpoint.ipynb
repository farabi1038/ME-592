```python
#CNN Network


class ConvNet(nn.Module):
    def __init__(self,num_classes=6):
        super(ConvNet,self).__init__()
        
        #Output size after convolution filter
        #((w-f+2P)/s) +1
        
        #Input shape= (256,3,150,150)
        #Input shape= (1,3,300,400)
        
        
        self.conv1=nn.Conv2d(in_channels=3,out_channels=12,kernel_size=3,stride=1,padding=1)
        #Shape= (256,12,150,150)
        #Shape= (1,12,300,400)
        
        
        self.bn1=nn.BatchNorm2d(num_features=12)
        #Shape= (256,12,150,150)
        #Shape= (1,12,300,400)
        self.relu1=nn.ReLU()
        #Shape= (256,12,150,150)
        #Shape= (1,12,300,400)
        
        
        self.pool=nn.MaxPool2d(kernel_size=2)
        #Reduce the image size be factor 2
        #Shape= (256,12,75,75)
        #Shape= (1,12,150,200)
        
        
        self.conv2=nn.Conv2d(in_channels=12,out_channels=20,kernel_size=3,stride=1,padding=1)
        #Shape= (256,20,75,75)
        self.relu2=nn.ReLU()
        #Shape= (256,20,75,75)
        #Shape= (1,20,150,200)
        
        
        
        
        self.conv3=nn.Conv2d(in_channels=20,out_channels=32,kernel_size=3,stride=1,padding=1)
        #Shape= (256,32,75,75)
        #Shape= (1,32,150,200)
        self.bn3=nn.BatchNorm2d(num_features=32)
        #Shape= (256,32,75,75)
        self.relu3=nn.ReLU()
        #Shape= (256,32,75,75)
        
        
        #self.fc=nn.Linear(in_features=150 * 200 * 32,out_features=num_classes)
        self.softmax=nn.Softmax()
        
        
        
        #Feed forwad function
        
    def forward(self,input):
        #print("input shape ",input.shape)
        output=self.conv1(input)
        output=self.bn1(output)
        output=self.relu1(output)
        #print("relu1 shape ",output.shape)
            
        output=self.pool(output)
        #print("pool shape ",output.shape)    
        output=self.conv2(output)
        output=self.relu2(output)
        #print("conv2 shape ",output.shape)
            
        output=self.conv3(output)
        output=self.bn3(output)
        output=self.relu3(output)
        #print("conv3 shape ",output.shape)    
            
            #Above output will be in matrix form, with shape (256,32,75,75)
        #print("hi",output.shape)    
        output=output.view(-1,32*150*200)
            
        #print("before fc",output.shape)     
        #output=self.fc(output)
        output=self.softmax(output)
            
        return output
```

```python
#checking for device
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model=ConvNet(num_classes=7).to(device)
```

```python
#Optmizer and loss function
optimizer=Adam(model.parameters(),lr=0.001,weight_decay=0.0001)
loss_function=nn.CrossEntropyLoss()
```

```python
num_epochs=10
train_count = 11781
test_count = 1472
print(train_count,test_count)
```

```python
#count=0
#for i, (images,labels) in enumerate(train_set):
#    print(labels)  
#images=images.unsqueeze(0)
#print(images.shape)
```

```python
#Model training and saving best model

best_accuracy=0.0

for epoch in range(num_epochs):
    
    #Evaluation and training on training dataset
    model.train()
    train_accuracy=0.0
    train_loss=0.0
    
    for i, (images,labels) in enumerate(trainLoader):
        if torch.cuda.is_available():
            images=Variable(images.cuda())
            labels=Variable(labels.cuda())
            
        optimizer.zero_grad()
        #images=images.unsqueeze(0)
        labels=torch.tensor(labels)
        #labels=labels.view (1)
        #print('before putting into model',images.shape)
        outputs=model(images)
        loss=loss_function(outputs,labels)
        loss.backward()
        optimizer.step()
        
        
        train_loss+= loss.cpu().data*images.size(0)
        _,prediction=torch.max(outputs.data,1)
        
        train_accuracy+=int(torch.sum(prediction==labels.data))
        
    train_accuracy=train_accuracy/train_count
    train_loss=train_loss/train_count
    
    
    # Evaluation on testing dataset
    model.eval()
    
    test_accuracy=0.0
    for i, (images,labels) in enumerate(testLoader):
        if torch.cuda.is_available():
            images=Variable(images.cuda())
            labels=Variable(labels.cuda())
        #images=images.unsqueeze(0)
        labels=torch.tensor(labels)
        #labels=labels.view (1)    
        outputs=model(images)
        
        _,prediction=torch.max(outputs.data,1)
        print("predicted value",prediction)
        print("label",labels.data)
        test_accuracy+=int(torch.sum(prediction==labels.data))
    
    test_accuracy=test_accuracy/test_count
    
    
    print('Epoch: '+str(epoch)+' Train Loss: '+str(train_loss)+' Train Accuracy: '+str(train_accuracy)+' Test Accuracy: '+str(test_accuracy))
    
    #Save the best model
    if test_accuracy>best_accuracy:
        torch.save(model.state_dict(),'best_checkpoint.model')
        best_accuracy=test_accuracy
    
```
